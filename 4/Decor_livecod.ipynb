# Декораторы

1. Время выполнения функции

```python
import time
```

```python
def measure_time(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        print(f'Время выполнения функции {func.__name__}: {end_time - start_time:.6f} секунд')
        return result
    return wrapper
```

```python
@measure_time
def slow_function(n):
    print('В процессе...')
    time.sleep(n)
    print('Готово!')

slow_function(3)
```

2. Логирование

```python
def log_func(func):
    def wrapper(*args, **kwargs):
        print(f'Вызов функции {func.__name__} с аргументами {args}, {kwargs}')
        result = func(*args, **kwargs)
        print(f'Функция {func.__name__} вернула {result}')
        return result
    return wrapper
```

```python
@log_func
def add(a, b):
    return a + b
```

```python
@log_func
def greet(name):
    return f'Привет, {name}!'
```

```python
add(5, b=12)
greet('Олег')
```

3. Декоратор с параметрами

```python
def check_access(role_required):
    def decor(func):
        def wrapper(user_role, *args, **kwargs):
            if user_role == role_required:
                return func(*args, **kwargs)
            else:
                print(f'Доступ к функции {func.__name__} запрещен для роли {user_role}.')
                return None
        return wrapper
    return decor
```

```python
@check_access('admin')
def del_user(user_id):
    return f'Пользователь с ID_{user_id} удален!'
```

```python
print(del_user('admin', 123))
```

```python
print(del_user('guest', 123))
```

4. Классовые декораторы (кэширование)

```python
def cache_result(func):
    cache = {}

    def wrapper(self, *args, **kwargs):
        if args in cache:
            print(f'Используем кэш для {func.__name__} с аргументами {args}')
            return cache[args]
        result = func(self, *args, **kwargs)
        cache[args] = result
        return result
    return wrapper
```

```python
class Calculator:
    @cache_result
    def add(self, a, b):
        print(f'Выполняю вычисление {a} + {b}')
        return a + b
    
    @cache_result
    def fib(self, n):
        if n == 0:
            return 0
        if n == 1:
            return 1
        return self.fib(n-1) + self.fib(n-2)
```

```python
calc = Calculator()
print(calc.add(7,5))
print(calc.add(7,5))
```

```python
start_time = time.time()
print(calc.fib(300))
print("Время выполнения:", time.time() - start_time)
```

```python
start_time = time.time()
print(calc.fib(300))
print("Время выполнения:", time.time() - start_time)
```

```python
start_time = time.time()
print(calc.fib(301))
print("Время выполнения:", time.time() - start_time)
```

```python
start_time = time.time()
print(calc.fib(301))
print("Время выполнения:", time.time() - start_time)
```

5. Вложенные декораторы
